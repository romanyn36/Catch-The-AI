# AI-Human Dataset
### this dataset extracted from conversations between AI and Human, mainly used for testing our models on real-world data
### The dataset contains 3 columns:
- Text: the text of the conversation
- label: the label of the conversation (AI or Human)
- model: the model that generated the conversation


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('AI-human dataset.xlsx')
df.head()
```

```python
labels = df['model'].value_counts()
plt.figure(figsize=(4, 4))
plt.pie(labels, labels=labels.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Text Sources')
plt.show()
```

```python

plt.figure(figsize=(8, 4))
sns.countplot(x='model', data=df)
plt.title('Count of Text by Source')
plt.xlabel('Source')
plt.ylabel('Count')
plt.show()
```

## length of text


```python
df['text_length'] = df['text'].apply(lambda x: len(x.split(' ')))
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='text_length', hue='model', multiple='stack', bins=30)
plt.title('Distribution of Text Length by Source')
plt.xlabel('Text Length')
plt.ylabel('Frequency')
plt.show()
```
